### Iterar sobre una lista

```python
# Utilizando la sentencia for()
mercado = ['Agua','Huevos','Aceite','Sal','Limon']
for producto in mercado:
    print(producto)
```

```python
# Iterar según su enumeración, enumerate()
for i, producto in enumerate(mercado):
    print(i, producto)
```

```python
# Iterar sobre múltiples listas, zip()
mercado = ['Agua','Aceite','Arroz']
marcas = ['Mineral','Oliva','Faraon']
for producto, marca in zip(mercado, marcas):
    print(producto, marca)
```

```python
# Nota: La función zip() realiza la combinación hasta que sea gota la lista más corta
mercado = ['Agua','Aceite','Arroz']
marcas = ['Mineral','Oliva','Faraon','Ferrari']
for producto, marca in zip(mercado, marcas):
    print(producto, marca)
```

### Listas por comprensión

```python
# Crear una lista de la forma clásica
valores = '32,45,11,87,20,48'
lista_de_valores = []
for valor in valores.split(','):
    valorentero = int(valor)
    lista_de_valores.append(valorentero)

print(lista_de_valores)
```

```python
# Crear una lista por comprensión
valores = '32,45,11,87,20,48'

lista_de_valores = [int(valor) for valor in valores.split(',')]
lista_de_valores
```
