```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
llamadas_recibidas = np.array([1000, 1110, 1110, 1240, 1100, 1200, 1050, 1030, 1050, 1090, 1130, 1150])
llamadas_atendidas = np.array([1000, 1105, 1110, 1250, 1100, 1200, 1050, 1000, 1020, 1040, 1060, 1070])
```

```python
dates = pd. date_range(start='2023-01-31', periods=12, freq='m')
dates
```

```python
dates_ = dates.strftime('%B').str[:3]
```

```python
plt.figure(figsize=(10,5), dpi = 90)
plt.plot(dates_, llamadas_recibidas, lw = 3, color = 'grey', marker = 'o')
plt.plot(dates_, llamadas_atendidas, lw = 3, color = 'royalblue', marker = 'o')
plt.axvline(6, ls = '--', color = 'grey')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.gca().spines['bottom'].set_color('grey')
plt.gca().spines['left'].set_color('grey')

plt.gca().tick_params(colors = '0.5', grid_color = 'grey')

plt.text(0.97, 0.61, 'Recibidas', transform = plt.gca().transAxes, color = 'gray', fontsize = 'x-large', fontweight = 'bold')
plt.text(0.97, 0.33, 'Atendidas', transform = plt.gca().transAxes, color = 'royalblue', fontsize = 'x-large', fontweight = 'bold')

plt.ylabel('Cantidad mensual de llamadas', color = 'grey')
plt.title('Comparacin de llamadas recibididas vs atendidas', color = 'grey', size = 12)

# Agregar etiquetas a partir de agosto
for i in range(7, len(dates)):
    plt.text(i + 0.15, llamadas_recibidas[i] + 20, llamadas_recibidas[i], color = 'gray', fontsize = 12, ha = 'center')
    plt.text(i + 0.15, llamadas_atendidas[i] - 25, llamadas_atendidas[i], color = 'royalblue', fontsize = 12, ha = 'center')

plt.ylim(900, 1300)
plt.xlim(0, 12)
```
