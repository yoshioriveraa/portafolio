```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
import math
```

```python
# 1.a
a = 0
b = 2*np.pi
espacio = 20
x = np.linspace(a, b, 360 // espacio + 1) # Van a ser 13 puntos para lograr equidistantes

f = np.sin(x)

plt.plot(x, f)
plt.plot(x, f, "o", color = "black")
plt.grid(True) #Rejillas
plt.show()
```

```python
# 1.b
inicio = -10
final = 10
puntos = 200
x = np.linspace(inicio, final, puntos)
f = np.power(x,2) + 5*x - 3
# Grfica
plt.title('Polinomio cuadrtico con 200 puntos')
plt.plot(x, f, color = "red", linestyle = "--", )
plt.grid(True)
plt.xlim(-10,10)
plt.ylim(min(f),max(x))
plt.show()
```

```python
# 1.c
inicio = 0
final = 6
x = np.linspace(inicio, final, 100)
f1 = 3*x*np.exp(x)
f2 = np.sin(x + 3)
# Grfica f1
plt.plot(x, f1, color = "blue", label = "F1")
# Grfica f2
plt.plot(x, f2, color = "red", linestyle = '--', label = "F2")
plt.legend()
plt.show()
```

```python
# 2.a

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
r = np.sqrt(x**2 + y**2)

f = 1/r

plt.plot(r, f)
plt.show()

```

```python
# 2.b
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
x = y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y) # Lista de matrices de Coordenadas
Z1 = 1 / np.sqrt(X**2 + (Y - 1)**2)
Z2 = 1 / np.sqrt(X**2 + (Y + 1)**2)
Z = Z1 - Z2
cs = ax.contour(X, Y, Z, 400, cmap = cm.coolwarm)
ax.clabel(cs, fontsize = 6)
plt.colorbar(cs)
plt.show()
```

```python
# 2.c
fig = plt.figure()
ax = fig.add_subplot(111)
x = y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = X - Y + 2
cs = ax.contour(X, Y, Z, 50, cmap = cm.coolwarm)
ax.clabel(cs, fontsize = 8)
plt.colorbar(cs)
plt.show()
```

```python
# 2.d
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
x = y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(16 - X**2 - Y**2)
cs = ax.contour(X, Y, Z, 200, cmap = cm.coolwarm)
ax.clabel(cs, fontsize = 6)
plt.colorbar(cs)
plt.show()
```

```python
# 2.e
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

x = y = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, y, )

Z = X**2 + Y**2 # z = 0
cs = ax.contour(X, Y, Z, 200, cmap = cm.coolwarm)
ax.clabel(cs, fontsize = 8)
plt.colorbar(cs)
plt.show()
```
